```python
# Sum ESG Scores
```

```python
hans_rosling.loc[:, 'rbics_sect_color'] = hans_rosling.loc[:, 'rbics_sect'].map(sect_color_dict)
```

```python
hans_rosling.loc[:, 'esg_combined'] = hans_rosling.loc[:, ['msci_esg_env', 'msci_esg_gov','msci_esg_social']].sum(axis=1)
```

# Building Hans Rosling Chart

```python
# Define a Function that accepts an End Date

def hans_rosling_date(year):
    # Define a function that filters the dataframe by date    
    date_filter = hans_rosling.loc[:, 'Date'] == year
    hans_rosling_year = hans_rosling.loc[date_filter, :]
    
    #arrays for scatter plot
    esg_score = hans_rosling_year.loc[:, 'esg_combined']
    sales = hans_rosling_year.loc[:, 'sales_total']
    mktval = hans_rosling_year.loc[:, 'mkt_val']
    rbics = hans_rosling_year.loc[:, 'rbics_econ']
    rbics_color = hans_rosling_year.loc[:, 'rbics_sect_color']
    ticker = hans_rosling_year.loc[:, 'Symbol']
    
    #labels for FAANG and other major companies
    #fb = hans_rosling_year.loc[:, 'Symbol'] == 'FB'
    #fb_label = hans_rosling_year.loc[fb, ['esg_combined', 'sales_total']]
    #fb_label_array = np.array(fb_label)

    #amzn = hans_rosling_year.loc[:, 'Symbol'] == 'AMZN'
    #amzn_label = hans_rosling_year.loc[amzn, ['esg_combined', 'sales_total']]
    #amzn_label_array = np.array(amzn_label)

    #aapl = hans_rosling_year.loc[:, 'Symbol'] == 'AAPL'
    #aapl_label = hans_rosling_year.loc[aapl, ['esg_combined', 'sales_total']]
    #aapl_label_array = np.array(aapl_label)

    #nflx = hans_rosling_year.loc[:, 'Symbol'] == 'NFLX'
    #nflx_label = hans_rosling_year.loc[nflx, ['esg_combined', 'sales_total']]
    #nflx_label_array = np.array(nflx_label)

    #goog = hans_rosling_year.loc[:, 'Symbol'] == 'GOOG'
    #goog_label = hans_rosling_year.loc[goog, ['esg_combined', 'sales_total']]
    #goog_label_array = np.array(goog_label)

    #msft = hans_rosling_year.loc[:, 'Symbol'] == 'MSFT'
    #msft_label = hans_rosling_year.loc[msft, ['esg_combined', 'sales_total']]
    #msft_label_array = np.array(msft_label)

    #brk = hans_rosling_year.loc[:, 'Symbol'] == 'BRK.B'
    #brk_label = hans_rosling_year.loc[brk, ['esg_combined', 'sales_total']]
    #brk_label_array = np.array(brk_label)

    #wmt = hans_rosling_year.loc[:, 'Symbol'] == 'WMT'
    #wmt_label = hans_rosling_year.loc[wmt, ['esg_combined', 'sales_total']]
    #wmt_label_array = np.array(wmt_label)

    #xom = hans_rosling_year.loc[:, 'Symbol'] == 'XOM'
    #xom_label = hans_rosling_year.loc[xom, ['esg_combined', 'sales_total']]
    #xom_label_array = np.array(xom_label)
    
    # Scatter plot
    plt.scatter(x = esg_score, y = sales, s = mktval/100 , edgecolors='black', c = rbics_color, alpha=.75)

    # customizations
    plt.yscale('log')
    plt.xlabel('ESG Score')
    plt.ylabel('Revenue')
    plt.xlim(0,25)
    plt.ylim(500, 700000)
    plt.title(year)
    #plt.xticks([1000,10000,100000], ['1k','10k','100k'])

    # Additional customizations
    #plt.text(fb_label_array[:, 0], fb_label_array[:, 1], 'FB')
    #plt.text(amzn_label_array[:, 0], amzn_label_array[:, 1], 'AMZN')
    #plt.text(aapl_label_array[:, 0], aapl_label_array[:, 1], 'AAPL')
    #plt.text(nflx_label_array[:, 0], nflx_label_array[:, 1], 'NFLX')
    #plt.text(goog_label_array[:, 0], goog_label_array[:, 1], 'GOOG')
    #plt.text(msft_label_array[:, 0], msft_label_array[:, 1], 'MSFT')
    #plt.text(brk_label_array[:, 0], brk_label_array[:, 1], 'BRK.B')
    #plt.text(wmt_label_array[:, 0], wmt_label_array[:, 1], 'WMT')
    #plt.text(xom_label_array[:, 0], xom_label_array[:, 1], 'XOM')

    # Add grid() call
    plt.grid(True)

    # return and display the plot
    return plt.show()

    
```
