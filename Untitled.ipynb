# Importing Data

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (16, 9)
plt.rcParams['font.size'] = 14

plt.style.use('fivethirtyeight')
```

```python
data  = Path('.', 'assets', 'rbics_esg_data.csv')
hans_rosling = pd.read_csv(data, na_values= '-')
```

# Cleaning Data

```python
hans_rosling = hans_rosling.dropna()
```

# Feature Engineering

```python
# create color map for 
# combine ESG data to one datapoint
```

```python
# Sect Color Map
# source:  https://matplotlib.org/2.0.2/examples/color/named_colors.html

sect_color_dict = {
    'Manufactured Products': 'black', 
    'Industrial Manufacturing': 'dimgray',
    'Healthcare Equipment': 'lightgray', 
    'Biopharmaceuticals': 'whitesmoke',
    'Software and Consulting': 'red', 
    'Miscellaneous Retail': 'crimson',
    'Electronic Components and Manufacturing': 'lightsalmon', 
    'Utilities': 'maroon',
    'Healthcare Services': 'rosybrown', 
    'Investment Services': 'saddlebrown', 
    'Insurance': 'chocolate',
    'Chemical, Plastic and Rubber Materials': 'darkorange', 
    'Real Estate': 'orange',
    'Mining and Mineral Products': 'navajowhite', 
    'Specialty Finance and Services': 'yellow',
    'Food and Tobacco Production': 'gold', 
    'Food and Staples Retail': 'lightgoldenrodyellow',
    'Industrial Services': 'lemonchiffon', 
    'Upstream Energy': 'green',
    'Downstream and Midstream Energy': 'lime', 
    'Hardware': 'purple',
    'Consumer Vehicles and Parts': 'seagreen', 
    'Telecommunications': 'cyan',
    'Business Services': 'navy', 
    'Banking': 'steelblue', 
    'Consumer Retail': 'skyblue',
    'Hospitality Services': 'lightblue', 
    'Media and Publishing Services': 'indigo',
    'Household Products': 'blueviolet', 
    'Consumer Goods': 'pink'
    
}
```

```python
# Sum ESG Scores
```

```python
hans_rosling.loc[:, 'rbics_sect_color'] = hans_rosling.loc[:, 'rbics_sect'].map(sect_color_dict)
```

```python
hans_rosling.loc[:, 'esg_combined'] = hans_rosling.loc[:, ['msci_esg_env', 'msci_esg_gov','msci_esg_social']].sum(axis=1)
```

# Chart for 2017

```python
#filter by Date for 2017
date = hans_rosling.loc[:, 'Date'] == '12/29/2017'
hans_rosling_2017 = hans_rosling.loc[date, :]
```

```python
hans_rosling_2017.head()
```

```python
# Analyze the impact of ESG Data, Market Cap, 
# and Revenue on S&P500 stocks against other companies in the same revere classification
```

```python
esg_score = hans_rosling_2017.loc[:, 'esg_combined']
sales = hans_rosling_2017.loc[:, 'sales_total']
mktval = hans_rosling_2017.loc[:, 'mkt_val']
rbics = hans_rosling_2017.loc[:, 'rbics_econ']
rbics_color = hans_rosling_2017.loc[:, 'rbics_sect_color']
ticker = hans_rosling_2017.loc[:, 'Symbol']
```

```python
rbics_color.unique()
```

```python
# Creating Labels for FAANG (Facebook, Amazon, Apple, Netflix, Google), MSFT Stocks
```

```python
#Scatter Plot for 2017

# Scatter plot
plt.scatter(x = esg_score, y = sales, s = mktval/100 , edgecolors='black', c = rbics_color, alpha=.75)

# Previous customizations
#plt.xscale('log')
plt.yscale('log')
plt.xlabel('ESG Score')
plt.ylabel('Revenue')
plt.title('2017')
#plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Additional customizations
plt.text(fb_label_array[:, 0], fb_label_array[:, 1], 'FB')
plt.text(amzn_label_array[:, 0], amzn_label_array[:, 1], 'AMZN')
plt.text(aapl_label_array[:, 0], aapl_label_array[:, 1], 'AAPL')
plt.text(nflx_label_array[:, 0], nflx_label_array[:, 1], 'NFLX')
plt.text(goog_label_array[:, 0], goog_label_array[:, 1], 'GOOG')
plt.text(msft_label_array[:, 0], msft_label_array[:, 1], 'MSFT')
plt.text(brk_label_array[:, 0], brk_label_array[:, 1], 'BRK.B')
plt.text(wmt_label_array[:, 0], wmt_label_array[:, 1], 'WMT')
plt.text(xom_label_array[:, 0], xom_label_array[:, 1], 'XOM')


# Add grid() call
plt.grid(True)



# Show the plot
plt.show()

```
